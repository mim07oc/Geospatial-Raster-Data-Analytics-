Algorithmic solution

```python
raster_2 = np.zeros(raster_size, dtype=np.uint8)
```

```python
# Define the "M" shape
for row in range(5, 15):  # Vertical lines of "M"
    raster_2[row, 5] = 1  # Left vertical line
    raster_2[row, 15] = 1  # Right vertical line

for i in range(6):  # Diagonals for the middle of "M"
    raster_2[5 + i, 5 + i] = 1  # Left diagonal of "M"
    raster_2[5 + i, 15 - i] = 1  # Right diagonal of "M"
```

```python
# Visualize the raster
plt.imshow(raster_2, cmap="gray", interpolation="nearest")
plt.title("Raster with Letter 'M'")
plt.axis("off")
plt.show()
```
